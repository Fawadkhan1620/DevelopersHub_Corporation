## About Dataset:

### - Iris Dataset Introduction

The Iris dataset is one of the most famous datasets in machine learning and statistics. It was introduced by Ronald A. Fisher in 1936 and is commonly used for classification tasks. The dataset contains measurements of iris flowers from three different species: Iris-setosa, Iris-versicolor, and Iris-virginica.

### - Objective

The main objective is to predict the species of an iris flower based on its sepal and petal measurements.

### - Column Description

- Id: Unique identifier assigned to each flower record. It is used only for identification and does not contribute to prediction.
- SepalLengthCm: Length of the sepal (outer leaf-like part of the flower) measured in centimeters.
- SepalWidthCm: Width of the sepal measured in centimeters.
- PetalLengthCm: Length of the petal (colored part of the flower) measured in centimeters.
- PetalWidthCm: Width of the petal measured in centimeters.
- Species: Target variable representing the species of the iris flower (Setosa, Versicolor, or Virginica).

### - Dataset Summary

- Dataset Type: Classification Dataset
- Target Variable: Species
- Input Features: SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm
- Number of Classes: 3 flower species
- Purpose: To classify iris flowers into their correct species based on flower measurements.

### - Overall Conclusion

The Iris dataset is a simple and well-balanced classification dataset that helps in understanding data preprocessing, visualization, feature analysis, and machine learning classification algorithms. Petal length and petal width are usually the most important features for distinguishing between different iris species.

```python
# Import all necassary libraries

import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

- warnings: ignore warnings.
- pandas: Pandas is used for data manipulation and analysis.
- matplotlib: Matplotlib is used for plotting graphs and charts.
- Seaborn: Seaborn is used for creating statistical visualizations.

```python
# Load and view first 3 rows of dataset

data = pd.read_csv('C:\\Users\\khan\\PycharmProjects\\pythonProject3\\Iris.csv')
data.head(3)
```

- data: store dataset in this variable.
- pd.read_csv: read mean display Iris dataset which is in csv (comma separated value) formate.
- head(3): show first 3 rows and if you want last rows use tail instead of head.

```python
# Structure of dataset

print('Number of rows and columns are: ', data.shape)
print()
print('Name of all columns are: ', data.columns)
```

- Shape: Sahpe is use to show total number of rows and columns (There are 150 rows & 6 columns).
- Columns: Columns is use to display all column names.

```python
# Scatter_plot between sepallength & petallength differ by species

sns.scatterplot(x=data['SepalLengthCm'], y=data['PetalLengthCm'], hue=data['Species'])
plt.grid()
plt.title('SepalLength & PetalLength By Species')
plt.show()


# Scatter_plot between sepalwidth & petalwidth differ by species

sns.scatterplot(x=data['SepalWidthCm'], y=data['PetalWidthCm'], hue=data['Species'])
plt.grid()
plt.title('SepalWidth & PetalWidth By Species')
plt.show()
```

- Scatter_plot: Show in the form of dots.
- 1st: First graph show Virginica has high length in cm.
- 2nd: Second graph show virginica has also high width in cm.

```python
# Histogram_plot (Shows the distribution of data)

plt.title('[All in one]')

plt.subplot(3,3,1)
sns.histplot(data['PetalLengthCm'], kde=True, color='b')
plt.title('PetalLength Dist')

plt.subplot(3,3,3)
sns.histplot(data['PetalWidthCm'], kde=True, color='y')
plt.title('PetalWidth Dist')

plt.subplot(3,3,7)
sns.histplot(data['SepalLengthCm'], kde=True, color='r')
plt.title('SepalLength Dist')

plt.subplot(3,3,9)
sns.histplot(data['SepalWidthCm'], kde=True, color='g')
plt.title('SepalWidth Dist')

plt.show()
```

- Subplot: Show all plots in one graph.
- Histogram: Show distribution of data also data are normally distributed or not.
- Bell_Curve: SepalWidthCm has bell curve so it's mean this is normally distributed.

```python
# Box_plot (Shows the min, max, median, quartiles and also outliers)

plt.subplot(2,3,1)
sns.boxplot(data['PetalLengthCm'], color='b')
plt.title('PetalLength')

plt.subplot(2,3,3)
sns.boxplot(data['PetalWidthCm'], color='y')
plt.title('PetalWidth')

plt.subplot(2,3,4)
sns.boxplot(data['SepalLengthCm'], color='r')
plt.title('SepalLength')

plt.subplot(2,3,6)
sns.boxplot(data['SepalWidthCm'], color='g')
plt.title('SepalWidth')

plt.show()
```

- Boxplot: Show minimum number, maximum number, median, quartiles and also outliers.
- SepalWidth: In sepalwidth graph there are some outliers show in the form of o shape.
