## About Dataset:

### - Insurance Dataset Introduction

The Medical Insurance Cost Prediction Dataset is a widely used dataset in data science and machine learning for predicting individual medical insurance expenses. It contains demographic, lifestyle, and health-related information about policyholders and their corresponding insurance charges.

The dataset was originally made popular through the Machine Learning with R book by Brett Lantz and is now commonly available on platforms such as Kaggle for educational and research purposes.

The primary goal of this dataset is to estimate a person's medical insurance claim amount (charges) based on factors such as age, gender, BMI, smoking habits, number of children, and residential region.

### - Objective

The objective of this project is to develop a Linear Regression model that predicts the medical insurance claim amount (charges) using personal and demographic information.

Specific objectives include:

1. Predict insurance charges using Linear Regression.
2. Analyze how Age affects insurance costs.

4. Analyze how smoking status affects insurance costs.
5. Evaluate model performance using:

### - Column Description

- age:	Age of the insurance policyholder in years.
- sex:	Gender of the individual (male or female).
- bmi:	Body Mass Index, calculated from height and weight. Higher BMI often indicates overweight or obesity.
- children:	Number of dependent children covered under the insurance plan.
- smoker:	Indicates whether the individual is a smoker (yes/no).
- region:	Residential area of the policyholder (northeast, northwest, southeast, southwest).
- charges:	Medical insurance cost charged to the individual. This is the target variable.

# Step 4: Visualization

```python
# BMI vs Charges

plt.figure(figsize=(8,5))
sns.scatterplot(x='bmi', y='charges', data=df)
plt.title("BMI vs Insurance Charges")
plt.show()


# Age vs Charges

plt.figure(figsize=(8,5))
sns.scatterplot(x='age', y='charges', data=df)
plt.title("Age vs Insurance Charges")
plt.show()


# Smoking Status vs Charges

plt.figure(figsize=(8,5))
sns.boxplot(x='smoker', y='charges', data=df)
plt.title("Smoking Status vs Insurance Charges")
plt.show()
```

# Step 5: Data Preprocessing

```python
# Convert categorical variables into numerical

le = LabelEncoder()

df['sex'] = le.fit_transform(df['sex'])
df['smoker'] = le.fit_transform(df['smoker'])
df['region'] = le.fit_transform(df['region'])

df.head(3)
```

- LabelEncoder: Encoder to encode the data.
- fit_transform: Fit encoder on this data and then transform it to new data.
- head(3): Agian show the first 3 rows of encoded dataset.

# Step 6: Feature Selection

```python
X = df.drop('charges', axis=1)
y = df['charges']
```

- X: Take all columns except charges column because this one is targeted column.
- y: Take only charges column.

# Step 7: Train-Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)
```

- train_test: Split data into training and testing parts.
- train_size: Always give train size greater than test size, I give here 80% data for training and the remaining will for testing. You also use test_size instead of train_size.
- random_state: Only for randomness, if you run again and again our model, So the result will not changing randomly again and again. You also use any other number instead of 42.

# Step 8: Train Linear Regression Model

```python
model = LinearRegression()

model.fit(X_train, y_train)
```

- model: Call the linear regression model and store in this variable.
- fit: Fit the model mean train the model on trainig data.

# Step 9: Predictions

```python
y_pred = model.predict(X_test)
```

- predict: Test the model prediction on testing data.

# Step 10: Model Evaluation

```python
mae = mean_absolute_error(y_test, y_pred)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))

r2 = r2_score(y_test, y_pred)

print("\nModel Performance")

print("MAE :", mae)
print("RMSE :", rmse)
print("R² Score :", r2)
```

- MAE, RMSE AND R2_SCORE: All are evaluation metrics, To check model result.

# Step 11: Actual vs Predicted

```python
plt.figure(figsize=(8,5))
plt.scatter(y_test, y_pred)

plt.xlabel("Actual Charges")
plt.ylabel("Predicted Charges")
plt.title("Actual vs Predicted Charges")

plt.show()
```

INTERPRETATION:

1. Mean Absolute Error (MAE) = 4186.51:
    - MAE measures the average difference between the actual insurance charges and the predicted charges.
2. Root Mean Squared Error (RMSE) = 5799.59:
    - RMSE measures the average prediction error while giving more weight to larger errors.
3. R² Score = 0.7833:
    - R² indicates how much of the variation in insurance charges is explained by the model.
